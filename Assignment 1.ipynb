```python
#clean tn_budgets to retain as much data as possible for comparison. Add an roi %, month and year column
```

```python
#rename movie to title
gross = tn_budgets.rename(columns = {'movie':'title'})
```

```python
#set index as title
gross = gross.set_index(['title'])
```

```python
#clean file to convert budget & gross to int
```

```python
#convert value to 
gross['domestic_gross'] = pd.to_numeric(gross['domestic_gross'].replace('[\$,]','', regex=True), errors='coerce')
```

```python
gross['production_budget'] = pd.to_numeric(gross['production_budget'].replace('[\$,]','', regex=True), errors='coerce')
```

```python
gross['worldwide_gross'] = pd.to_numeric(gross['worldwide_gross'].replace('[\$,]','', regex=True), errors='coerce')
```

```python
#change release_date to datetime and add year and month column
gross['release_date'] = pd.to_datetime(gross['release_date'])

#add year and month columns
gross['year'] = gross['release_date'].dt.year
gross['month'] = gross['release_date'].dt.month
```

```python
gross.head()
```

```python
gross.groupby(['month']).mean()
```

```python
#format to dollars
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
gross.groupby(['month']).mean()
```

```python
#remove columns id
gross = gross.drop(columns=['id'])
```

```python
#create ROI column and profit column
gross['profit'] = gross['worldwide_gross'] - gross['production_budget']
gross['roi %'] = gross['profit'] / gross['production_budget'] * 100
```

```python
gross.head()
```

```python
gross.describe()
```

```python
gross.drop_duplicates(keep='first')
```

```python
#remove 0 values in gross revenue and budget
gross = gross.loc[gross['worldwide_gross'] * gross['production_budget'] != 0]
```

```python
gross.info()
```

```python
#clean for primary and secondary genre
#clean tmdb_movies dataframe, remove uneeded columns
clean_tmdb_movies = tmdb_movies.drop(columns=['Unnamed: 0', 'id', 'original_language', 'original_title', 'release_date'])
```

```python
clean_tmdb_movies.info()
```

```python
#set 'title' as index
clean_tmdb_movies = clean_tmdb_movies.set_index('title')
```

```python
clean_tmdb_movies.head()
```

```python
#split first value from genre_ids string to get primary genre
clean_tmdb_movies['primary_genre_id'] = clean_tmdb_movies['genre_ids'].str.split(n=1).str[0]
```

```python
#clean to have only the id number
clean_tmdb_movies['primary_genre_id'] = clean_tmdb_movies['primary_genre_id'].str.replace('[','')
clean_tmdb_movies['primary_genre_id'] = clean_tmdb_movies['primary_genre_id'].str.replace(',','')
```

```python
#map genres to numbers as found on tmdb website
genres = {'28':'Action',
          '12':'Adventure',
          '16':'Animation',
          '35':'Comedy',
          '80':'Crime',
          '99':'Documentary',
          '18':'Drama',
          '10751':'Family',
          '14':'Fantasy',
          '36':'History',
          '27':'Horror',
          '10402':'Music',
          '9648':'Mystery',
          '10749':'Romance',
          '878':'Science Fiction',
          '10770':'TV Movie',
          '53':'Thriller',
          '10752':'War',
          '37':'Western'}

clean_tmdb_movies['primary_genre'] = clean_tmdb_movies['primary_genre_id'].map(genres)
```

```python
#check unique values of genre
clean_tmdb_movies['primary_genre'].nunique()
```

```python
#split second value from genre_ids string to get primary genre
clean_tmdb_movies['secondary_genre_id'] = clean_tmdb_movies['genre_ids'].str.split(n=2).str[1]
clean_tmdb_movies['secondary_genre_id'] = clean_tmdb_movies['secondary_genre_id'].str.replace(',','')
```

```python
#map genres to seconday genre id
clean_tmdb_movies['secondary_genre'] = clean_tmdb_movies['secondary_genre_id'].map(genres)
```

```python
clean_tmdb_movies.head()
```

```python
clean_tmdb_movies = clean_tmdb_movies.drop(columns=['genre_ids', 'primary_genre_id', 'secondary_genre_id'])
```

```python
clean_tmdb_movies.head()
```

```python
clean_tmdb_movies.drop_duplicates(keep='first')
```

```python
#merge genre and gross df's
gross_v_genre = gross.merge(clean_tmdb_movies, on = 'title', how = 'inner')
gross_v_genre.info()
```

```python
gross_v_genre.drop_duplicates(keep='first')
```

```python
gross_v_genre.groupby(['primary_genre']).mean(numeric_only=True)
```

```python
# prepare dataframe

gross_v_genre_2 = gross_v_genre[gross_v_genre.year > 2010]
gross_v_genre_2 = gross_v_genre_2.groupby(['year','primary_genre'])['roi %'].mean()

# unstack the `continent` index, to place it as columns
gross_v_genre_2 = gross_v_genre_2.unstack(level='primary_genre')

# The name of columns would become the name of legend
# when using dataframe plot
gross_v_genre_2.columns.name = 'roi %'

# Now, we have a 2d talbe, 1st column become to X
# and other columns become to Y
# In [14]: df.head()
# Out[14]:
# Life Expectation     Africa  Americas       Asia     Europe  Oceania
# year
# 1952              39.135500  53.27984  46.314394  64.408500   69.255
# 1957              41.266346  55.96028  49.318544  66.703067   70.295
# 1962              43.319442  58.39876  51.563223  68.539233   71.085
# 1967              45.334538  60.41092  54.663640  69.737600   71.310
# 1972              47.450942  62.39492  57.319269  70.775033   71.910

# matplotlib operations
# Here we use dataframe plot function
# You could also use raw matplotlib plot one column each to do fine control
# Please polish the figure with more configurations
fig, ax = plt.subplots(figsize=(6, 4.5))
gross_v_genre_2.plot()
```
